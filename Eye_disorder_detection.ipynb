```python
#converting npy normalized to normalised image dataset


import os
import numpy as np
import cv2
from tqdm import tqdm

# =========================
# PATHS
# =========================
NPY_ROOT = "MyDriveBackup/eyediseases_dataset1_CLEANED/Preprocessing/Normalized Dataset"
PNG_ROOT = "MyDriveBackup/eyediseases_dataset1_CLEANED/Preprocessing/Normalized_PNG"          # output .png files

os.makedirs(PNG_ROOT, exist_ok=True)

# =========================
# FUNCTION: normalize to [0,255]
# =========================
def normalize_to_uint8(img):
    """
    Converts normalized or standardized image to uint8 [0,255]
    """
    img = img.astype(np.float32)

    # If image is standardized (mean-std)
    if img.min() < 0 or img.max() > 1:
        img = (img - img.min()) / (img.max() - img.min() + 1e-8)

    img = (img * 255.0).clip(0, 255)
    return img.astype(np.uint8)

# =========================
# CONVERSION LOOP
# =========================
classes = os.listdir(NPY_ROOT)

for cls in tqdm(classes, desc="Converting Classes"):
    cls_npy_path = os.path.join(NPY_ROOT, cls)
    cls_png_path = os.path.join(PNG_ROOT, cls)
    os.makedirs(cls_png_path, exist_ok=True)

    npy_files = [f for f in os.listdir(cls_npy_path) if f.endswith(".npy")]

    for file in tqdm(npy_files, desc=f"{cls}", leave=False):
        npy_path = os.path.join(cls_npy_path, file)

        # Load npy
        img = np.load(npy_path)

        # Handle grayscale → RGB
        if img.ndim == 2:
            img = np.stack([img]*3, axis=-1)

        # Normalize safely
        img = normalize_to_uint8(img)

        # Save as PNG
        png_name = file.replace(".npy", ".png")
        png_path = os.path.join(cls_png_path, png_name)

        cv2.imwrite(png_path, cv2.cvtColor(img, cv2.COLOR_RGB2BGR))

print("✅ Conversion completed: .npy → normalized .png")
```

```python
# resnet50 without descriptor

# ==========================================================
# STEP 0: IMPORTS
# ==========================================================
import os
import numpy as np

import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.layers import (
    Input, GlobalAveragePooling2D,
    Dense, Dropout
)
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, classification_report,
    cohen_kappa_score, roc_auc_score
)

import matplotlib.pyplot as plt
import seaborn as sns

# ==========================================================
# STEP 1: PATHS & PARAMETERS
# ==========================================================

TRAIN_DIR = "MyDriveBackup/eyediseases_dataset1_CLEANED/Test-train-split/train"
VAL_DIR   = "MyDriveBackup/eyediseases_dataset1_CLEANED/Test-train-split/val"

RESULTS_DIR = "MyDriveBackup/eyediseases_dataset1_CLEANED/resnet50_baseline_results"
os.makedirs(RESULTS_DIR, exist_ok=True)

IMG_HEIGHT = 224
IMG_WIDTH  = 224
BATCH_SIZE = 16
NUM_CLASSES = 5
EPOCHS = 25

# ==========================================================
# STEP 2: DATA GENERATORS
# ==========================================================

train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    zoom_range=0.1
)

val_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=True
)

val_generator = val_datagen.flow_from_directory(
    VAL_DIR,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=False
)

# ==========================================================
# STEP 3: RESNET50 BASELINE MODEL (NO CAPD)
# ==========================================================

def ResNet50_baseline(input_shape=(224,224,3), num_classes=5):

    inputs = Input(shape=input_shape)

    base_model = ResNet50(
        weights="imagenet",
        include_top=False,
        input_tensor=inputs
    )

    # Freeze backbone
    for layer in base_model.layers:
        layer.trainable = False

    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(512, activation="relu")(x)
    x = Dropout(0.5)(x)
    outputs = Dense(num_classes, activation="softmax")(x)

    return Model(inputs, outputs)

# ==========================================================
# STEP 4: COMPILE MODEL
# ==========================================================

model = ResNet50_baseline(
    input_shape=(IMG_HEIGHT, IMG_WIDTH, 3),
    num_classes=NUM_CLASSES
)

model.compile(
    optimizer=tf.keras.optimizers.SGD(
        learning_rate=0.001, momentum=0.9
    ),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# ==========================================================
# STEP 5: CALLBACKS
# ==========================================================

checkpoint_cb = tf.keras.callbacks.ModelCheckpoint(
    os.path.join(RESULTS_DIR, "ResNet50_baseline_best.keras"),
    monitor="val_accuracy",
    save_best_only=True
)

earlystop_cb = tf.keras.callbacks.EarlyStopping(
    monitor="val_accuracy",
    patience=5,
    restore_best_weights=True
)

# ==========================================================
# STEP 6: TRAINING
# ==========================================================

history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=val_generator,
    callbacks=[checkpoint_cb, earlystop_cb],
    verbose=1
)

model.save(os.path.join(RESULTS_DIR, "ResNet50_baseline_final.keras"))

# ==========================================================
# STEP 7: EVALUATION
# ==========================================================

val_generator.reset()
y_prob = model.predict(val_generator, verbose=1)
y_pred = np.argmax(y_prob, axis=1)
y_true = val_generator.classes
class_names = list(val_generator.class_indices.keys())

acc = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average="macro")
recall = recall_score(y_true, y_pred, average="macro")
f1 = f1_score(y_true, y_pred, average="macro")
kappa = cohen_kappa_score(y_true, y_pred)

auc = roc_auc_score(
    tf.keras.utils.to_categorical(y_true, NUM_CLASSES),
    y_prob,
    average="macro",
    multi_class="ovr"
)

# ==========================================================
# STEP 8: SAVE METRICS
# ==========================================================

with open(os.path.join(RESULTS_DIR, "metrics.txt"), "w") as f:
    f.write(f"Accuracy    : {acc:.4f}\n")
    f.write(f"Precision   : {precision:.4f}\n")
    f.write(f"Recall      : {recall:.4f}\n")
    f.write(f"F1-score    : {f1:.4f}\n")
    f.write(f"Cohen Kappa : {kappa:.4f}\n")
    f.write(f"AUC         : {auc:.4f}\n")

# ==========================================================
# STEP 9: CONFUSION MATRIX
# ==========================================================

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(
    cm, annot=True, fmt="d", cmap="Blues",
    xticklabels=class_names,
    yticklabels=class_names
)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix – ResNet50 Baseline")
plt.tight_layout()
plt.savefig(
    os.path.join(RESULTS_DIR, "confusion_matrix.png"),
    dpi=300
)
plt.close()

# ==========================================================
# STEP 10: CLASSIFICATION REPORT
# ==========================================================

report = classification_report(
    y_true, y_pred, target_names=class_names
)

with open(
    os.path.join(RESULTS_DIR, "classification_report.txt"), "w"
) as f:
    f.write(report)

print("🎉 Training & evaluation completed successfully!")
print(f"📁 Results saved at: {RESULTS_DIR}")

```

```python
# train -test split and restnet 50 on capd

# ==========================================================
# STEP 0: IMPORTS
# ==========================================================
import os
import shutil
import random
import numpy as np
from tqdm import tqdm

import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.layers import (
    Input, Conv2D, GlobalAveragePooling2D,
    Dense, Multiply, Concatenate, Dropout
)
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, classification_report,
    cohen_kappa_score, roc_auc_score
)

import matplotlib.pyplot as plt
import seaborn as sns

# ==========================================================
# STEP 1: PATHS & PARAMETERS
# ==========================================================

SOURCE_DIR = "MyDriveBackup/eyediseases_dataset1_CLEANED/Preprocessing/Normalized_PNG"
SPLIT_DIR  = "MyDriveBackup/eyediseases_dataset1_CLEANED/Test-train-split"

TRAIN_DIR = os.path.join(SPLIT_DIR, "train")
VAL_DIR   = os.path.join(SPLIT_DIR, "val")

# UPDATED: results folder inside the dataset folder
RESULTS_DIR = "MyDriveBackup/eyediseases_dataset1_CLEANED/resnet_capd_results"
os.makedirs(RESULTS_DIR, exist_ok=True)

IMG_HEIGHT = 224
IMG_WIDTH  = 224
BATCH_SIZE = 16
NUM_CLASSES = 5
EPOCHS = 25
SPLIT_RATIO = 0.8
SEED = 42

# ==========================================================
# STEP 2: TRAIN–VALIDATION SPLIT (ONE TIME)
# ==========================================================

if not os.path.exists(SPLIT_DIR):

    print("🔹 Creating train / val split from normalized images...")

    classes = sorted(os.listdir(SOURCE_DIR))
    random.seed(SEED)

    for cls in tqdm(classes, desc="Splitting classes"):
        cls_path = os.path.join(SOURCE_DIR, cls)
        images = os.listdir(cls_path)
        random.shuffle(images)

        split_idx = int(len(images) * SPLIT_RATIO)
        train_imgs = images[:split_idx]
        val_imgs   = images[split_idx:]

        os.makedirs(os.path.join(TRAIN_DIR, cls), exist_ok=True)
        os.makedirs(os.path.join(VAL_DIR, cls), exist_ok=True)

        for img in train_imgs:
            shutil.copy(os.path.join(cls_path, img),
                        os.path.join(TRAIN_DIR, cls, img))

        for img in val_imgs:
            shutil.copy(os.path.join(cls_path, img),
                        os.path.join(VAL_DIR, cls, img))

    print("✅ Train/Validation split completed.")

else:
    print("✅ Train/Val split already exists.")

# ==========================================================
# STEP 3: DATA GENERATORS (IMAGENET COMPATIBLE)
# ==========================================================

train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    zoom_range=0.1
)

val_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=True
)

val_generator = val_datagen.flow_from_directory(
    VAL_DIR,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=False
)

# ==========================================================
# STEP 4: CHANNEL ATTENTION (CAPD)
# ==========================================================

def channel_attention(feature, ratio=8):
    channels = feature.shape[-1]
    avg_pool = GlobalAveragePooling2D()(feature)
    avg_pool = Dense(channels // ratio, activation="relu")(avg_pool)
    avg_pool = Dense(channels, activation="sigmoid")(avg_pool)
    return Multiply()([feature, avg_pool])

# ==========================================================
# STEP 5: CAPD-RESNET50 MODEL
# ==========================================================

def CAPD_ResNet50(input_shape=(224,224,3), num_classes=5):

    inputs = Input(shape=input_shape)

    base_model = ResNet50(
        weights="imagenet",
        include_top=False,
        input_tensor=inputs
    )

    F1 = base_model.get_layer("conv2_block3_out").output
    F2 = base_model.get_layer("conv3_block4_out").output
    F3 = base_model.get_layer("conv4_block6_out").output

    F1 = channel_attention(F1)
    F2 = channel_attention(F2)
    F3 = channel_attention(F3)

    F1 = Conv2D(256, (1,1), activation="relu")(F1)
    F2 = Conv2D(256, (1,1), activation="relu")(F2)
    F3 = Conv2D(256, (1,1), activation="relu")(F3)

    F1 = GlobalAveragePooling2D()(F1)
    F2 = GlobalAveragePooling2D()(F2)
    F3 = GlobalAveragePooling2D()(F3)

    descriptor = Concatenate(name="CAPD_Descriptor")([F1, F2, F3])

    x = Dense(512, activation="relu")(descriptor)
    x = Dropout(0.5)(x)
    outputs = Dense(num_classes, activation="softmax")(x)

    return Model(inputs, outputs)

# ==========================================================
# STEP 6: COMPILE MODEL
# ==========================================================

model = CAPD_ResNet50()

model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.001, momentum=0.9),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# ==========================================================
# STEP 7: CALLBACKS
# ==========================================================

checkpoint_cb = tf.keras.callbacks.ModelCheckpoint(
    os.path.join(RESULTS_DIR, "ResNet50_CAPD_best.keras"),
    monitor="val_accuracy",
    save_best_only=True
)

earlystop_cb = tf.keras.callbacks.EarlyStopping(
    monitor="val_accuracy",
    patience=5,
    restore_best_weights=True
)

# ==========================================================
# STEP 8: TRAINING
# ==========================================================

history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=val_generator,
    callbacks=[checkpoint_cb, earlystop_cb],
    verbose=1
)

model.save(os.path.join(RESULTS_DIR, "ResNet50_CAPD_final.keras"))

# ==========================================================
# STEP 9: EVALUATION
# ==========================================================

val_generator.reset()
y_prob = model.predict(val_generator, verbose=1)
y_pred = np.argmax(y_prob, axis=1)
y_true = val_generator.classes
class_names = list(val_generator.class_indices.keys())

acc = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average="macro")
recall = recall_score(y_true, y_pred, average="macro")
f1 = f1_score(y_true, y_pred, average="macro")
kappa = cohen_kappa_score(y_true, y_pred)

auc = roc_auc_score(
    tf.keras.utils.to_categorical(y_true, NUM_CLASSES),
    y_prob,
    average="macro",
    multi_class="ovr"
)

# ==========================================================
# STEP 10: SAVE METRICS
# ==========================================================

with open(os.path.join(RESULTS_DIR, "metrics.txt"), "w") as f:
    f.write(f"Accuracy      : {acc:.4f}\n")
    f.write(f"Precision     : {precision:.4f}\n")
    f.write(f"Recall        : {recall:.4f}\n")
    f.write(f"F1-score      : {f1:.4f}\n")
    f.write(f"Cohen Kappa   : {kappa:.4f}\n")
    f.write(f"AUC           : {auc:.4f}\n")

# ==========================================================
# STEP 11: CONFUSION MATRIX
# ==========================================================

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(
    cm, annot=True, fmt="d", cmap="Blues",
    xticklabels=class_names,
    yticklabels=class_names
)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix – ResNet50 + CAPD")
plt.tight_layout()
plt.savefig(os.path.join(RESULTS_DIR, "confusion_matrix.png"), dpi=300)
plt.close()

# ==========================================================
# STEP 12: CLASSIFICATION REPORT
# ==========================================================

report = classification_report(y_true, y_pred, target_names=class_names)
with open(os.path.join(RESULTS_DIR, "classification_report.txt"), "w") as f:
    f.write(report)

print("🎉 Training & evaluation completed successfully!")
print(f"📁 All results saved in folder: {RESULTS_DIR}")
```

```python
#hyperparameter tuning in bayesian part 1 25 epoch

# ==========================================================
# STEP 0: IMPORTS
# ==========================================================
import os
import numpy as np
import tensorflow as tf
import keras_tuner as kt
import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.layers import (
    Input, Conv2D, GlobalAveragePooling2D,
    Dense, Multiply, Concatenate, Dropout
)
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report,
    cohen_kappa_score, roc_auc_score, roc_curve
)

# ==========================================================
# STEP 1: PATHS & PARAMETERS
# ==========================================================

BASE_DIR = "MyDriveBackup/eyediseases_dataset1_CLEANED/Test-train-split"
TRAIN_DIR = os.path.join(BASE_DIR, "train")
VAL_DIR   = os.path.join(BASE_DIR, "val")

RESULTS_DIR = "MyDriveBackup/eyediseases_dataset1_CLEANED/resnet_capd_bayesian_results"
os.makedirs(RESULTS_DIR, exist_ok=True)

IMG_SIZE = (224, 224)
BATCH_SIZE = 16
NUM_CLASSES = 5
FINAL_EPOCHS = 25

# ==========================================================
# STEP 2: DATA GENERATORS
# ==========================================================

train_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=20,
    horizontal_flip=True,
    zoom_range=0.1
)

val_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

train_data = train_gen.flow_from_directory(
    TRAIN_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=True
)

val_data = val_gen.flow_from_directory(
    VAL_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=False
)

# ==========================================================
# STEP 3: CAPD MODULE
# ==========================================================

def channel_attention(x, ratio=8):
    ch = x.shape[-1]
    gap = GlobalAveragePooling2D()(x)
    d1 = Dense(ch // ratio, activation="relu")(gap)
    d2 = Dense(ch, activation="sigmoid")(d1)
    return Multiply()([x, d2])

# ==========================================================
# STEP 4: MODEL BUILDER (BAYESIAN)
# ==========================================================

def build_capd_model(hp):

    inputs = Input(shape=(224,224,3))
    base = ResNet50(weights="imagenet", include_top=False, input_tensor=inputs)

    for layer in base.layers:
        layer.trainable = False

    F1 = channel_attention(base.get_layer("conv2_block3_out").output)
    F2 = channel_attention(base.get_layer("conv3_block4_out").output)
    F3 = channel_attention(base.get_layer("conv4_block6_out").output)

    F1 = GlobalAveragePooling2D()(Conv2D(256,(1,1),activation="relu")(F1))
    F2 = GlobalAveragePooling2D()(Conv2D(256,(1,1),activation="relu")(F2))
    F3 = GlobalAveragePooling2D()(Conv2D(256,(1,1),activation="relu")(F3))

    descriptor = Concatenate()([F1, F2, F3])

    x = Dense(
        hp.Int("dense_units", 256, 1024, step=256),
        activation="relu"
    )(descriptor)

    x = Dropout(hp.Float("dropout", 0.3, 0.6, step=0.1))(x)
    outputs = Dense(NUM_CLASSES, activation="softmax")(x)

    model = Model(inputs, outputs)

    model.compile(
        optimizer=tf.keras.optimizers.SGD(
            learning_rate=hp.Float("lr", 1e-4, 1e-2, sampling="log"),
            momentum=0.9
        ),
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model

# ==========================================================
# STEP 5: BAYESIAN TUNING
# ==========================================================

tuner = kt.BayesianOptimization(
    build_capd_model,
    objective="val_accuracy",
    max_trials=10,
    directory=RESULTS_DIR,
    project_name="CAPD_Bayesian"
)

tuner.search(train_data, validation_data=val_data, epochs=10)

best_model = tuner.get_best_models(1)[0]

# ==========================================================
# STEP 6: FINAL TRAINING
# ==========================================================

history = best_model.fit(
    train_data,
    validation_data=val_data,
    epochs=FINAL_EPOCHS,
    verbose=1
)

best_model.save(os.path.join(RESULTS_DIR, "ResNet50_CAPD_Bayesian_best.keras"))

# ==========================================================
# STEP 7: EVALUATION METRICS
# ==========================================================

val_data.reset()
y_prob = best_model.predict(val_data)
y_pred = np.argmax(y_prob, axis=1)
y_true = val_data.classes
class_names = list(val_data.class_indices.keys())

acc = accuracy_score(y_true, y_pred)
prec = precision_score(y_true, y_pred, average="macro", zero_division=0)
rec  = recall_score(y_true, y_pred, average="macro", zero_division=0)
f1   = f1_score(y_true, y_pred, average="macro", zero_division=0)
kappa = cohen_kappa_score(y_true, y_pred)

auc = roc_auc_score(
    tf.keras.utils.to_categorical(y_true, NUM_CLASSES),
    y_prob,
    multi_class="ovr",
    average="macro"
)

with open(os.path.join(RESULTS_DIR, "metrics.txt"), "w") as f:
    f.write(
        f"Accuracy: {acc:.4f}\nPrecision: {prec:.4f}\nRecall: {rec:.4f}\n"
        f"F1-score: {f1:.4f}\nKappa: {kappa:.4f}\nAUC: {auc:.4f}\n"
    )

# ==========================================================
# STEP 8: CONFUSION MATRIX
# ==========================================================

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d",
            xticklabels=class_names,
            yticklabels=class_names, cmap="Blues")
plt.title("Confusion Matrix – CAPD")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.savefig(os.path.join(RESULTS_DIR, "confusion_matrix.png"), dpi=300)
plt.close()

# ==========================================================
# STEP 9: ROC CURVE
# ==========================================================

plt.figure()
for i in range(NUM_CLASSES):
    fpr, tpr, _ = roc_curve(
        tf.keras.utils.to_categorical(y_true, NUM_CLASSES)[:, i],
        y_prob[:, i]
    )
    plt.plot(fpr, tpr, label=f"{class_names[i]}")

plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves – CAPD")
plt.legend()
plt.savefig(os.path.join(RESULTS_DIR, "roc_curves.png"), dpi=300)
plt.close()

# ==========================================================
# STEP 10: TRAINING & TESTING CURVES
# ==========================================================

epochs = range(1, len(history.history["accuracy"]) + 1)

plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
plt.plot(epochs, history.history["accuracy"], label="Train Acc")
plt.plot(epochs, history.history["val_accuracy"], label="Val Acc")
plt.xlabel("Epochs"); plt.ylabel("Accuracy"); plt.legend(); plt.grid()

plt.subplot(1,2,2)
plt.plot(epochs, history.history["loss"], label="Train Loss")
plt.plot(epochs, history.history["val_loss"], label="Val Loss")
plt.xlabel("Epochs"); plt.ylabel("Loss"); plt.legend(); plt.grid()

plt.tight_layout()
plt.savefig(os.path.join(RESULTS_DIR, "training_testing_curves.png"), dpi=300)
plt.close()

print("✅ CAPD + Bayesian Optimization: All results saved successfully!")
```

```python
# restnet50 for rhd feature extraction


# ==========================================================
# STEP 0: IMPORTS
# ==========================================================
import os
import numpy as np
from tqdm import tqdm

import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.layers import (
    Input, Conv2D, GlobalAveragePooling2D,
    Dense, Concatenate, Dropout, Add
)
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, classification_report,
    cohen_kappa_score, roc_auc_score
)

import matplotlib.pyplot as plt
import seaborn as sns

# ==========================================================
# STEP 1: PATHS & PARAMETERS
# ==========================================================

BASE_DIR = "MyDriveBackup/eyediseases_dataset1_CLEANED/Test-train-split"

TRAIN_DIR = os.path.join(BASE_DIR, "train")
VAL_DIR   = os.path.join(BASE_DIR, "val")

RESULTS_DIR = "MyDriveBackup/eyediseases_dataset1_CLEANED/resnet_rhd_results"
os.makedirs(RESULTS_DIR, exist_ok=True)

IMG_HEIGHT = 224
IMG_WIDTH  = 224
BATCH_SIZE = 16
NUM_CLASSES = 5
EPOCHS = 25

# ==========================================================
# STEP 2: DATA GENERATORS
# ==========================================================

train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    zoom_range=0.1
)

val_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=True
)

val_generator = val_datagen.flow_from_directory(
    VAL_DIR,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=False
)

# ==========================================================
# STEP 3: RHD – Residual Hierarchical Descriptor
# ==========================================================

def residual_block(x, filters):
    shortcut = x

    # Project shortcut if channel mismatch
    if x.shape[-1] != filters:
        shortcut = Conv2D(filters, (1,1), padding="same")(shortcut)

    x = Conv2D(filters, (1,1), activation="relu", padding="same")(x)
    x = Conv2D(filters, (3,3), activation="relu", padding="same")(x)

    return Add()([x, shortcut])


# ==========================================================
# STEP 4: RHD-RESNET50 MODEL
# ==========================================================

def RHD_ResNet50(input_shape=(224,224,3), num_classes=5):

    inputs = Input(shape=input_shape)

    base_model = ResNet50(
        weights="imagenet",
        include_top=False,
        input_tensor=inputs
    )

    # Hierarchical feature extraction
    F1 = base_model.get_layer("conv2_block3_out").output
    F2 = base_model.get_layer("conv3_block4_out").output
    F3 = base_model.get_layer("conv4_block6_out").output

    # Residual refinement
    F1 = residual_block(F1, 256)
    F2 = residual_block(F2, 256)
    F3 = residual_block(F3, 256)

    # Global pooling
    F1 = GlobalAveragePooling2D()(F1)
    F2 = GlobalAveragePooling2D()(F2)
    F3 = GlobalAveragePooling2D()(F3)

    # RHD descriptor
    descriptor = Concatenate(name="RHD_Descriptor")([F1, F2, F3])

    x = Dense(512, activation="relu")(descriptor)
    x = Dropout(0.5)(x)
    outputs = Dense(num_classes, activation="softmax")(x)

    return Model(inputs, outputs)

# ==========================================================
# STEP 5: COMPILE MODEL
# ==========================================================

model = RHD_ResNet50()

model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.001, momentum=0.9),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# ==========================================================
# STEP 6: CALLBACKS
# ==========================================================

checkpoint_cb = tf.keras.callbacks.ModelCheckpoint(
    os.path.join(RESULTS_DIR, "ResNet50_RHD_best.keras"),
    monitor="val_accuracy",
    save_best_only=True
)

earlystop_cb = tf.keras.callbacks.EarlyStopping(
    monitor="val_accuracy",
    patience=5,
    restore_best_weights=True
)

# ==========================================================
# STEP 7: TRAINING
# ==========================================================

history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=val_generator,
    callbacks=[checkpoint_cb, earlystop_cb],
    verbose=1
)

model.save(os.path.join(RESULTS_DIR, "ResNet50_RHD_final.keras"))

# ==========================================================
# STEP 8: EVALUATION
# ==========================================================

val_generator.reset()
y_prob = model.predict(val_generator, verbose=1)


# ==========================================================
# STEP 8: EVALUATION METRICS
# ==========================================================

# Reset generator
val_generator.reset()

# Predictions
y_prob = model.predict(val_generator, verbose=1)
y_pred = np.argmax(y_prob, axis=1)
y_true = val_generator.classes
class_names = list(val_generator.class_indices.keys())

# Metrics
acc = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average="macro", zero_division=0)
recall = recall_score(y_true, y_pred, average="macro", zero_division=0)
f1 = f1_score(y_true, y_pred, average="macro", zero_division=0)
kappa = cohen_kappa_score(y_true, y_pred)

auc = roc_auc_score(
    tf.keras.utils.to_categorical(y_true, NUM_CLASSES),
    y_prob,
    average="macro",
    multi_class="ovr"
)

# Print metrics
print("\n📊 Validation Performance")
print(f"Accuracy      : {acc:.4f}")
print(f"Precision     : {precision:.4f}")
print(f"Recall        : {recall:.4f}")
print(f"F1-score      : {f1:.4f}")
print(f"Cohen Kappa   : {kappa:.4f}")
print(f"AUC           : {auc:.4f}")

# ==========================================================
# STEP 9: SAVE METRICS TO FILE
# ==========================================================

with open(os.path.join(RESULTS_DIR, "validation_metrics.txt"), "w") as f:
    f.write(f"Accuracy      : {acc:.4f}\n")
    f.write(f"Precision     : {precision:.4f}\n")
    f.write(f"Recall        : {recall:.4f}\n")
    f.write(f"F1-score      : {f1:.4f}\n")
    f.write(f"Cohen Kappa   : {kappa:.4f}\n")
    f.write(f"AUC           : {auc:.4f}\n")

# ==========================================================
# STEP 10: CONFUSION MATRIX
# ==========================================================

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=class_names,
    yticklabels=class_names
)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix – ResNet50 + RHD")
plt.tight_layout()
plt.savefig(os.path.join(RESULTS_DIR, "confusion_matrix.png"), dpi=300)
plt.close()

# ==========================================================
# STEP 11: CLASSIFICATION REPORT
# ==========================================================

report = classification_report(y_true, y_pred, target_names=class_names)
with open(os.path.join(RESULTS_DIR, "classification_report.txt"), "w") as f:
    f.write(report)

print("\n✅ Evaluation completed successfully!")
print(f"📁 Results saved in: {RESULTS_DIR}")




```

```python
# hyper parameter tuning in Bayesian part 1 epoch 10

# ==========================================================
# STEP 0: IMPORTS
# ==========================================================
import os
import numpy as np
import tensorflow as tf
import keras_tuner as kt
import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.layers import (
    Input, Conv2D, GlobalAveragePooling2D,
    Dense, Concatenate, Dropout, Add
)
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report,
    cohen_kappa_score, roc_auc_score, roc_curve
)

# ==========================================================
# STEP 1: PATHS & PARAMETERS
# ==========================================================

BASE_DIR = "MyDriveBackup/eyediseases_dataset1_CLEANED/Test-train-split"
TRAIN_DIR = os.path.join(BASE_DIR, "train")
VAL_DIR   = os.path.join(BASE_DIR, "val")

RESULTS_DIR = "MyDriveBackup/eyediseases_dataset1_CLEANED/resnet_rhd_bayesian_results"
os.makedirs(RESULTS_DIR, exist_ok=True)

IMG_SIZE = (224, 224)
BATCH_SIZE = 16
NUM_CLASSES = 5
FINAL_EPOCHS = 25   # 🔹 Part-1 requirement

# ==========================================================
# STEP 2: DATA GENERATORS
# ==========================================================

train_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=20,
    horizontal_flip=True,
    zoom_range=0.1
)

val_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

train_data = train_gen.flow_from_directory(
    TRAIN_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=True
)

val_data = val_gen.flow_from_directory(
    VAL_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=False
)

# ==========================================================
# STEP 3: RHD MODULE (Residual Hierarchical Descriptor)
# ==========================================================

def residual_block(x, filters):
    shortcut = Conv2D(filters, (1,1), padding="same")(x)
    x = Conv2D(filters, (1,1), activation="relu", padding="same")(x)
    x = Conv2D(filters, (3,3), activation="relu", padding="same")(x)
    return Add()([x, shortcut])

# ==========================================================
# STEP 4: MODEL BUILDER (BAYESIAN – RHD)
# ==========================================================

def build_rhd_model(hp):

    inputs = Input(shape=(224,224,3))
    base = ResNet50(weights="imagenet", include_top=False, input_tensor=inputs)

    # Freeze backbone
    for layer in base.layers:
        layer.trainable = False

    # Hierarchical feature extraction
    F1 = base.get_layer("conv2_block3_out").output
    F2 = base.get_layer("conv3_block4_out").output
    F3 = base.get_layer("conv4_block6_out").output

    # Residual hierarchical refinement
    F1 = residual_block(F1, 256)
    F2 = residual_block(F2, 256)
    F3 = residual_block(F3, 256)

    # Global pooling
    F1 = GlobalAveragePooling2D()(F1)
    F2 = GlobalAveragePooling2D()(F2)
    F3 = GlobalAveragePooling2D()(F3)

    # RHD descriptor
    descriptor = Concatenate(name="RHD_Descriptor")([F1, F2, F3])

    x = Dense(
        hp.Int("dense_units", 256, 1024, step=256),
        activation="relu"
    )(descriptor)

    x = Dropout(
        hp.Float("dropout", 0.3, 0.6, step=0.1)
    )(x)

    outputs = Dense(NUM_CLASSES, activation="softmax")(x)

    model = Model(inputs, outputs)

    model.compile(
        optimizer=tf.keras.optimizers.SGD(
            learning_rate=hp.Float("lr", 1e-4, 1e-2, sampling="log"),
            momentum=0.9
        ),
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model

# ==========================================================
# STEP 5: BAYESIAN HYPERPARAMETER TUNING
# ==========================================================

tuner = kt.BayesianOptimization(
    build_rhd_model,
    objective="val_accuracy",
    max_trials=10,
    directory=RESULTS_DIR,
    project_name="RHD_Bayesian"
)

tuner.search(
    train_data,
    validation_data=val_data,
    epochs=10    # 🔹 short tuning phase
)

best_model = tuner.get_best_models(1)[0]

# ==========================================================
# STEP 6: FINAL TRAINING (25 EPOCHS)
# ==========================================================

history = best_model.fit(
    train_data,
    validation_data=val_data,
    epochs=FINAL_EPOCHS,
    verbose=1
)

best_model.save(
    os.path.join(RESULTS_DIR, "ResNet50_RHD_Bayesian_best.keras")
)

# ==========================================================
# STEP 7: EVALUATION METRICS
# ==========================================================

val_data.reset()
y_prob = best_model.predict(val_data)
y_pred = np.argmax(y_prob, axis=1)
y_true = val_data.classes
class_names = list(val_data.class_indices.keys())

acc = accuracy_score(y_true, y_pred)
prec = precision_score(y_true, y_pred, average="macro", zero_division=0)
rec  = recall_score(y_true, y_pred, average="macro", zero_division=0)
f1   = f1_score(y_true, y_pred, average="macro", zero_division=0)
kappa = cohen_kappa_score(y_true, y_pred)

auc = roc_auc_score(
    tf.keras.utils.to_categorical(y_true, NUM_CLASSES),
    y_prob,
    multi_class="ovr",
    average="macro"
)

with open(os.path.join(RESULTS_DIR, "metrics.txt"), "w") as f:
    f.write(
        f"Accuracy: {acc:.4f}\nPrecision: {prec:.4f}\nRecall: {rec:.4f}\n"
        f"F1-score: {f1:.4f}\nKappa: {kappa:.4f}\nAUC: {auc:.4f}\n"
    )

    # ==========================================================
# STEP 7.1: CLASSIFICATION REPORT (RHD)
# ==========================================================

report = classification_report(
    y_true,
    y_pred,
    target_names=class_names,
    digits=4
)

print("\n📋 Classification Report – RHD:\n")
print(report)

with open(os.path.join(RESULTS_DIR, "classification_report.txt"), "w") as f:
    f.write(report)

# ==========================================================
# STEP 8: CONFUSION MATRIX
# ==========================================================

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(
    cm, annot=True, fmt="d",
    xticklabels=class_names,
    yticklabels=class_names, cmap="Blues"
)
plt.title("Confusion Matrix – RHD")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.savefig(os.path.join(RESULTS_DIR, "confusion_matrix.png"), dpi=300)
plt.close()

# ==========================================================
# STEP 9: ROC CURVES
# ==========================================================

plt.figure()
for i in range(NUM_CLASSES):
    fpr, tpr, _ = roc_curve(
        tf.keras.utils.to_categorical(y_true, NUM_CLASSES)[:, i],
        y_prob[:, i]
    )
    plt.plot(fpr, tpr, label=class_names[i])

plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves – RHD")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(RESULTS_DIR, "roc_curves.png"), dpi=300)
plt.close()

# ==========================================================
# STEP 10: TRAINING & TESTING CURVES
# ==========================================================

epochs = range(1, len(history.history["accuracy"]) + 1)

plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
plt.plot(epochs, history.history["accuracy"], label="Training Accuracy")
plt.plot(epochs, history.history["val_accuracy"], label="Validation Accuracy")
plt.xlabel("Epochs"); plt.ylabel("Accuracy"); plt.legend(); plt.grid()

plt.subplot(1,2,2)
plt.plot(epochs, history.history["loss"], label="Training Loss")
plt.plot(epochs, history.history["val_loss"], label="Validation Loss")
plt.xlabel("Epochs"); plt.ylabel("Loss"); plt.legend(); plt.grid()

plt.tight_layout()
plt.savefig(os.path.join(RESULTS_DIR, "training_testing_curves.png"), dpi=300)
plt.close()

print("🎉 RHD + Bayesian Hyperparameter Tuning (Part-1) COMPLETED!")


```

```python
#hyperparameter tuning on bayesian of ssslfd_block

# ==========================================================
# STEP 0: IMPORTS
# ==========================================================
import os
import numpy as np
import tensorflow as tf
import keras_tuner as kt
import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.layers import (
    Input, Conv2D, GlobalAveragePooling2D,
    Dense, Concatenate, Dropout
)
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report,
    cohen_kappa_score, roc_auc_score, roc_curve
)

# ==========================================================
# STEP 1: PATHS & PARAMETERS
# ==========================================================

BASE_DIR = "MyDriveBackup/eyediseases_dataset1_CLEANED/Test-train-split"
TRAIN_DIR = os.path.join(BASE_DIR, "train")
VAL_DIR   = os.path.join(BASE_DIR, "val")

RESULTS_DIR = "MyDriveBackup/eyediseases_dataset1_CLEANED/resnet_sslfd_bayesian_results"
os.makedirs(RESULTS_DIR, exist_ok=True)

IMG_SIZE = (224, 224)
BATCH_SIZE = 16
NUM_CLASSES = 5
FINAL_EPOCHS = 25   # Part-1 requirement

# ==========================================================
# STEP 2: DATA GENERATORS
# ==========================================================

train_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=20,
    horizontal_flip=True,
    zoom_range=0.1
)

val_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

train_data = train_gen.flow_from_directory(
    TRAIN_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=True
)

val_data = val_gen.flow_from_directory(
    VAL_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=False
)

# ==========================================================
# STEP 3: SSLFD MODULE
# (Feature Distillation / Self-Similarity based CNN Descriptor)
# ==========================================================

def sslfd_block(x, filters=256):
    """
    SSLFD block:
    - Projects features into a shared embedding space
    - Enables local feature consistency across layers
    """
    x = Conv2D(filters, (1,1), activation="relu", padding="same")(x)
    x = Conv2D(filters, (3,3), activation="relu", padding="same")(x)
    return x

# ==========================================================
# STEP 4: MODEL BUILDER (BAYESIAN – SSLFD)
# ==========================================================

def build_sslfd_model(hp):

    inputs = Input(shape=(224,224,3))
    base = ResNet50(weights="imagenet", include_top=False, input_tensor=inputs)

    # Freeze backbone
    for layer in base.layers:
        layer.trainable = False

    # Multi-level feature extraction
    F1 = base.get_layer("conv2_block3_out").output
    F2 = base.get_layer("conv3_block4_out").output
    F3 = base.get_layer("conv4_block6_out").output

    # SSLFD refinement
    F1 = sslfd_block(F1, 256)
    F2 = sslfd_block(F2, 256)
    F3 = sslfd_block(F3, 256)

    # Global pooling
    F1 = GlobalAveragePooling2D()(F1)
    F2 = GlobalAveragePooling2D()(F2)
    F3 = GlobalAveragePooling2D()(F3)

    # SSLFD Descriptor
    descriptor = Concatenate(name="SSLFD_Descriptor")([F1, F2, F3])

    x = Dense(
        hp.Int("dense_units", 256, 1024, step=256),
        activation="relu"
    )(descriptor)

    x = Dropout(
        hp.Float("dropout", 0.3, 0.6, step=0.1)
    )(x)

    outputs = Dense(NUM_CLASSES, activation="softmax")(x)

    model = Model(inputs, outputs)

    model.compile(
        optimizer=tf.keras.optimizers.SGD(
            learning_rate=hp.Float("lr", 1e-4, 1e-2, sampling="log"),
            momentum=0.9
        ),
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model

# ==========================================================
# STEP 5: BAYESIAN HYPERPARAMETER TUNING
# ==========================================================

tuner = kt.BayesianOptimization(
    build_sslfd_model,
    objective="val_accuracy",
    max_trials=10,
    directory=RESULTS_DIR,
    project_name="SSLFD_Bayesian"
)

tuner.search(
    train_data,
    validation_data=val_data,
    epochs=10   # short tuning phase
)

best_model = tuner.get_best_models(1)[0]

# ==========================================================
# STEP 6: FINAL TRAINING (25 EPOCHS)
# ==========================================================

history = best_model.fit(
    train_data,
    validation_data=val_data,
    epochs=FINAL_EPOCHS,
    verbose=1
)

best_model.save(
    os.path.join(RESULTS_DIR, "ResNet50_SSLFD_Bayesian_best.keras")
)

# ==========================================================
# STEP 7: EVALUATION METRICS
# ==========================================================

val_data.reset()
y_prob = best_model.predict(val_data)
y_pred = np.argmax(y_prob, axis=1)
y_true = val_data.classes
class_names = list(val_data.class_indices.keys())

acc = accuracy_score(y_true, y_pred)
prec = precision_score(y_true, y_pred, average="macro", zero_division=0)
rec  = recall_score(y_true, y_pred, average="macro", zero_division=0)
f1   = f1_score(y_true, y_pred, average="macro", zero_division=0)
kappa = cohen_kappa_score(y_true, y_pred)

auc = roc_auc_score(
    tf.keras.utils.to_categorical(y_true, NUM_CLASSES),
    y_prob,
    multi_class="ovr",
    average="macro"
)

with open(os.path.join(RESULTS_DIR, "metrics.txt"), "w") as f:
    f.write(
        f"Accuracy: {acc:.4f}\nPrecision: {prec:.4f}\nRecall: {rec:.4f}\n"
        f"F1-score: {f1:.4f}\nKappa: {kappa:.4f}\nAUC: {auc:.4f}\n"
    )

# ==========================================================
# STEP 8: CLASSIFICATION REPORT
# ==========================================================

report = classification_report(
    y_true, y_pred, target_names=class_names, digits=4
)

print("\n📋 Classification Report – SSLFD:\n")
print(report)

with open(os.path.join(RESULTS_DIR, "classification_report.txt"), "w") as f:
    f.write(report)

# ==========================================================
# STEP 9: CONFUSION MATRIX
# ==========================================================

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(
    cm, annot=True, fmt="d",
    xticklabels=class_names,
    yticklabels=class_names, cmap="Blues"
)
plt.title("Confusion Matrix – SSLFD")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.savefig(os.path.join(RESULTS_DIR, "confusion_matrix.png"), dpi=300)
plt.close()

# ==========================================================
# STEP 10: ROC CURVES
# ==========================================================

plt.figure()
for i in range(NUM_CLASSES):
    fpr, tpr, _ = roc_curve(
        tf.keras.utils.to_categorical(y_true, NUM_CLASSES)[:, i],
        y_prob[:, i]
    )
    plt.plot(fpr, tpr, label=class_names[i])

plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves – SSLFD")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(RESULTS_DIR, "roc_curves.png"), dpi=300)
plt.close()

# ==========================================================
# STEP 11: TRAINING & TESTING CURVES
# ==========================================================

epochs = range(1, len(history.history["accuracy"]) + 1)

plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
plt.plot(epochs, history.history["accuracy"], label="Training Accuracy")
plt.plot(epochs, history.history["val_accuracy"], label="Validation Accuracy")
plt.xlabel("Epochs"); plt.ylabel("Accuracy"); plt.legend(); plt.grid()

plt.subplot(1,2,2)
plt.plot(epochs, history.history["loss"], label="Training Loss")
plt.plot(epochs, history.history["val_loss"], label="Validation Loss")
plt.xlabel("Epochs"); plt.ylabel("Loss"); plt.legend(); plt.grid()

plt.tight_layout()
plt.savefig(os.path.join(RESULTS_DIR, "training_testing_curves.png"), dpi=300)
plt.close()

print("🎉 SSLFD + Bayesian Hyperparameter Tuning (Part-1) COMPLETED!")
```

```python
#weighted fusion

# ==========================================================
# WEIGHTED FEATURE FUSION (FINAL COMPLETE CODE)
# ==========================================================

import os
import numpy as np

# ==========================================================
# PATHS
# ==========================================================

BASE_DIR = "MyDriveBackup/eyediseases_dataset1_CLEANED"

CAPD_DIR  = os.path.join(BASE_DIR, "resnet_capd_bayesian_results")
RHD_DIR   = os.path.join(BASE_DIR, "resnet_rhd_bayesian_results")
SSLFD_DIR = os.path.join(BASE_DIR, "resnet_sslfd_bayesian_results")

# Feature files
CAPD_FEATURES  = os.path.join(CAPD_DIR, "capd_features.npy")
RHD_FEATURES   = os.path.join(RHD_DIR, "rhd_features.npy")
SSLFD_FEATURES = os.path.join(SSLFD_DIR, "sslfd_features.npy")

# Metrics files
CAPD_METRICS  = os.path.join(CAPD_DIR, "metrics.txt")
RHD_METRICS   = os.path.join(RHD_DIR, "metrics.txt")
SSLFD_METRICS = os.path.join(SSLFD_DIR, "metrics.txt")

# Output
FUSED_SAVE_PATH = os.path.join(BASE_DIR, "weighted_fused_features.npy")

# ==========================================================
# FUNCTION: READ ACCURACY FROM METRICS FILE
# ==========================================================

def read_accuracy(metrics_file):
    with open(metrics_file, "r") as f:
        for line in f:
            if "Accuracy" in line:
                return float(line.split(":")[1].strip())
    raise ValueError(f"Accuracy not found in {metrics_file}")

# ==========================================================
# STEP 1: LOAD FEATURES
# ==========================================================

F_capd  = np.load(CAPD_FEATURES)
F_rhd   = np.load(RHD_FEATURES)
F_sslfd = np.load(SSLFD_FEATURES)

print("Feature shapes:")
print("CAPD :", F_capd.shape)
print("RHD  :", F_rhd.shape)
print("SSLFD:", F_sslfd.shape)

# Sanity check
assert F_capd.shape == F_rhd.shape == F_sslfd.shape, "❌ Feature shape mismatch!"

# ==========================================================
# STEP 2: READ REAL VALIDATION ACCURACIES
# ==========================================================

acc_capd  = read_accuracy(CAPD_METRICS)
acc_rhd   = read_accuracy(RHD_METRICS)
acc_sslfd = read_accuracy(SSLFD_METRICS)

print("\nValidation Accuracies:")
print("CAPD :", acc_capd)
print("RHD  :", acc_rhd)
print("SSLFD:", acc_sslfd)

# ==========================================================
# STEP 3: COMPUTE NORMALIZED WEIGHTS
# ==========================================================

total_acc = acc_capd + acc_rhd + acc_sslfd

w_capd  = acc_capd  / total_acc
w_rhd   = acc_rhd   / total_acc
w_sslfd = acc_sslfd / total_acc

print("\nFusion Weights:")
print("CAPD :", w_capd)
print("RHD  :", w_rhd)
print("SSLFD:", w_sslfd)

# ==========================================================
# STEP 4: WEIGHTED FEATURE FUSION
# ==========================================================

F_fused = (
    w_capd  * F_capd +
    w_rhd   * F_rhd +
    w_sslfd * F_sslfd
)

print("\nFused feature shape:", F_fused.shape)

# ==========================================================
# STEP 5: SAVE FUSED FEATURES
# ==========================================================

np.save(FUSED_SAVE_PATH, F_fused)

print("\n✅ WEIGHTED FEATURE FUSION COMPLETED SUCCESSFULLY!")
print(f"📁 Fused features saved at:\n{FUSED_SAVE_PATH}")
```

```python
# ==========================================================
# DESCRIPTOR-LEVEL FUSED METRICS (CAPD + RHD + SSLFD)
# ==========================================================

import os

# ==========================================================
# PATHS
# ==========================================================

BASE_DIR = "MyDriveBackup/eyediseases_dataset1_CLEANED"

CAPD_METRICS  = os.path.join(BASE_DIR, "resnet_capd_bayesian_results", "metrics.txt")
RHD_METRICS   = os.path.join(BASE_DIR, "resnet_rhd_bayesian_results", "metrics.txt")
SSLFD_METRICS = os.path.join(BASE_DIR, "resnet_sslfd_bayesian_results", "metrics.txt")

SAVE_PATH = os.path.join(BASE_DIR, "fused_metrics.txt")

# ==========================================================
# FUNCTION: READ METRIC FROM metrics.txt
# ==========================================================

def read_metric(metrics_file, metric_name):
    with open(metrics_file, "r") as f:
        for line in f:
            if metric_name.lower() in line.lower():
                return float(line.split(":")[1].strip())
    raise ValueError(f"{metric_name} not found in {metrics_file}")

# ==========================================================
# STEP 1: READ METRICS FOR EACH DESCRIPTOR
# ==========================================================

# ---- CAPD ----
A_capd = read_metric(CAPD_METRICS, "Accuracy")
P_capd = read_metric(CAPD_METRICS, "Precision")
R_capd = read_metric(CAPD_METRICS, "Recall")
F_capd = read_metric(CAPD_METRICS, "F1")

# ---- RHD ----
A_rhd = read_metric(RHD_METRICS, "Accuracy")
P_rhd = read_metric(RHD_METRICS, "Precision")
R_rhd = read_metric(RHD_METRICS, "Recall")
F_rhd = read_metric(RHD_METRICS, "F1")

# ---- SSLFD ----
A_sslfd = read_metric(SSLFD_METRICS, "Accuracy")
P_sslfd = read_metric(SSLFD_METRICS, "Precision")
R_sslfd = read_metric(SSLFD_METRICS, "Recall")
F_sslfd = read_metric(SSLFD_METRICS, "F1")

print("\n🔹 Individual Descriptor Metrics Loaded Successfully")

# ==========================================================
# STEP 2: COMPUTE ACCURACY-BASED WEIGHTS
# ==========================================================

total_acc = A_capd + A_rhd + A_sslfd

w_capd  = A_capd  / total_acc
w_rhd   = A_rhd   / total_acc
w_sslfd = A_sslfd / total_acc

print("\n🔹 Fusion Weights")
print(f"CAPD  : {w_capd:.4f}")
print(f"RHD   : {w_rhd:.4f}")
print(f"SSLFD : {w_sslfd:.4f}")

# ==========================================================
# STEP 3: DESCRIPTOR-LEVEL FUSED METRICS
# ==========================================================

fused_accuracy  = w_capd*A_capd  + w_rhd*A_rhd  + w_sslfd*A_sslfd
fused_precision = w_capd*P_capd  + w_rhd*P_rhd  + w_sslfd*P_sslfd
fused_recall    = w_capd*R_capd  + w_rhd*R_rhd  + w_sslfd*R_sslfd
fused_f1        = w_capd*F_capd  + w_rhd*F_rhd  + w_sslfd*F_sslfd

# ==========================================================
# STEP 4: PRINT RESULTS
# ==========================================================

print("\n✅ DESCRIPTOR-LEVEL FUSED METRICS")
print(f"Fused Accuracy  : {fused_accuracy:.4f}")
print(f"Fused Precision : {fused_precision:.4f}")
print(f"Fused Recall    : {fused_recall:.4f}")
print(f"Fused F1-score  : {fused_f1:.4f}")

# ==========================================================
# STEP 5: SAVE RESULTS
# ==========================================================

with open(SAVE_PATH, "w") as f:
    f.write("Descriptor-level fused metrics (accuracy-weighted aggregation)\n\n")
    f.write(f"CAPD Accuracy  : {A_capd:.4f}\n")
    f.write(f"RHD Accuracy   : {A_rhd:.4f}\n")
    f.write(f"SSLFD Accuracy : {A_sslfd:.4f}\n\n")

    f.write("Fusion Weights\n")
    f.write(f"CAPD  : {w_capd:.4f}\n")
    f.write(f"RHD   : {w_rhd:.4f}\n")
    f.write(f"SSLFD : {w_sslfd:.4f}\n\n")

    f.write("Fused Metrics\n")
    f.write(f"Accuracy  : {fused_accuracy:.4f}\n")
    f.write(f"Precision : {fused_precision:.4f}\n")
    f.write(f"Recall    : {fused_recall:.4f}\n")
    f.write(f"F1-score  : {fused_f1:.4f}\n")

print(f"\n📁 Fused metrics saved at:\n{SAVE_PATH}")
print("\n🎉 DESCRIPTOR-LEVEL FUSION COMPLETED SUCCESSFULLY!")
```

```python
# ==========================================================
# HIGHEST THRESHOLD SELECTION AFTER KPCA (SINGLE BLOCK)
# ==========================================================

import numpy as np

# ----------------------------------------------------------
# PATH
# ----------------------------------------------------------

KPCA_FEATURES_PATH = (
    "MyDriveBackup/eyediseases_dataset1_CLEANED/kpca_rbf_features.npy"
)

# ----------------------------------------------------------
# STEP 1: LOAD KPCA FEATURES
# ----------------------------------------------------------

F_kpca = np.load(KPCA_FEATURES_PATH)
print("Loaded KPCA features shape:", F_kpca.shape)   # (1115, 128)

# ----------------------------------------------------------
# STEP 2: CONFIDENCE SCORE CALCULATION
# (L2 norm of KPCA features)
# ----------------------------------------------------------

confidence_scores = np.linalg.norm(F_kpca, axis=1)

print("Confidence scores shape:", confidence_scores.shape)

# ----------------------------------------------------------
# STEP 3: HIGHEST THRESHOLD SELECTION
# ----------------------------------------------------------

highest_threshold = np.max(confidence_scores)
threshold_95 = np.percentile(confidence_scores, 95)
threshold_90 = np.percentile(confidence_scores, 90)

print("\nThresholds:")
print("Highest Threshold (Max):", highest_threshold)
print("95% Threshold:", threshold_95)
print("90% Threshold:", threshold_90)

# ----------------------------------------------------------
# STEP 4: FINAL SELECTION USING BEST THRESHOLD
# (Recommended: 95% percentile)
# ----------------------------------------------------------

selected_idx = confidence_scores >= threshold_95

print("\nSelected samples (above threshold):", np.sum(selected_idx))
print("Rejected / low-confidence samples:", np.sum(~selected_idx))

# ----------------------------------------------------------
# STEP 5: SAVE THRESHOLD FOR DEPLOYMENT
# ----------------------------------------------------------

np.save(
    "MyDriveBackup/eyediseases_dataset1_CLEANED/final_threshold.npy",
    threshold_95
)

print("\n✅ Highest-confidence threshold saved successfully!")
```

```python
# ==========================================================
# FINAL SOFTMAX CLASSIFIER AFTER KPCA (UPDATED & SIMPLE)
# ==========================================================

import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.resnet50 import preprocess_input

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report,
    cohen_kappa_score, roc_curve, auc
)

# ==========================================================
# PATHS
# ==========================================================

BASE_DIR = "MyDriveBackup/eyediseases_dataset1_CLEANED"

KPCA_FEATURES_PATH = os.path.join(BASE_DIR, "kpca_rbf_features.npy")
THRESHOLD_PATH = os.path.join(BASE_DIR, "final_threshold.npy")

VAL_DIR = os.path.join(BASE_DIR, "Test-train-split/val")

RESULTS_DIR = os.path.join(BASE_DIR, "final_softmax_results")
os.makedirs(RESULTS_DIR, exist_ok=True)

NUM_CLASSES = 5
EPOCHS = 50
BATCH_SIZE = 8
IMG_SIZE = (224, 224)

# ==========================================================
# STEP 1: LOAD KPCA FEATURES & THRESHOLD
# ==========================================================

X = np.load(KPCA_FEATURES_PATH)
threshold = np.load(THRESHOLD_PATH)

print("KPCA features shape:", X.shape)
print("Threshold:", threshold)

# ==========================================================
# STEP 2: LOAD LABELS DIRECTLY FROM VAL FOLDER ✅
# ==========================================================

val_gen = ImageDataGenerator(preprocessing_function=preprocess_input)

val_data = val_gen.flow_from_directory(
    VAL_DIR,
    target_size=IMG_SIZE,
    batch_size=1,
    class_mode="categorical",
    shuffle=False
)

y = val_data.classes
class_names = list(val_data.class_indices.keys())

print("Labels shape:", y.shape)
print("Classes:", class_names)

# ==========================================================
# STEP 3: APPLY THRESHOLD SELECTION
# ==========================================================

confidence_scores = np.linalg.norm(X, axis=1)
selected_idx = confidence_scores >= threshold

X_sel = X[selected_idx]
y_sel = y[selected_idx]

print("Selected samples:", X_sel.shape[0])

y_sel_cat = to_categorical(y_sel, NUM_CLASSES)

# ==========================================================
# STEP 4: SOFTMAX CLASSIFIER
# ==========================================================

model = Sequential([
    Dense(128, activation="relu", input_shape=(X_sel.shape[1],)),
    Dropout(0.4),
    Dense(NUM_CLASSES, activation="softmax")
])

model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# ==========================================================
# STEP 5: TRAINING
# ==========================================================

history = model.fit(
    X_sel, y_sel_cat,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    verbose=1
)

model.save(os.path.join(RESULTS_DIR, "final_softmax_model.keras"))

# ==========================================================
# STEP 6: PREDICTION
# ==========================================================

y_prob = model.predict(X_sel)
y_pred = np.argmax(y_prob, axis=1)

# ==========================================================
# STEP 7: METRICS
# ==========================================================

acc = accuracy_score(y_sel, y_pred)
prec = precision_score(y_sel, y_pred, average="macro", zero_division=0)
rec = recall_score(y_sel, y_pred, average="macro", zero_division=0)
f1 = f1_score(y_sel, y_pred, average="macro", zero_division=0)
kappa = cohen_kappa_score(y_sel, y_pred)

# ROC & AUC
y_true_oh = to_categorical(y_sel, NUM_CLASSES)
fpr, tpr, roc_auc = {}, {}, {}

for i in range(NUM_CLASSES):
    fpr[i], tpr[i], _ = roc_curve(y_true_oh[:, i], y_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

macro_auc = np.mean(list(roc_auc.values()))

# ==========================================================
# STEP 8: SAVE METRICS
# ==========================================================

with open(os.path.join(RESULTS_DIR, "metrics.txt"), "w") as f:
    f.write(f"Accuracy  : {acc:.4f}\n")
    f.write(f"Precision : {prec:.4f}\n")
    f.write(f"Recall    : {rec:.4f}\n")
    f.write(f"F1-score  : {f1:.4f}\n")
    f.write(f"Kappa     : {kappa:.4f}\n")
    f.write(f"Macro AUC : {macro_auc:.4f}\n")

print("✅ Metrics saved")

# ==========================================================
# STEP 9: CONFUSION MATRIX
# ==========================================================

cm = confusion_matrix(y_sel, y_pred)

plt.figure(figsize=(7,6))
sns.heatmap(
    cm, annot=True, fmt="d", cmap="Blues",
    xticklabels=class_names,
    yticklabels=class_names
)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix – Final Softmax")
plt.savefig(os.path.join(RESULTS_DIR, "confusion_matrix.png"), dpi=300)
plt.close()

# ==========================================================
# STEP 10: ROC CURVES
# ==========================================================

plt.figure()
for i in range(NUM_CLASSES):
    plt.plot(
        fpr[i], tpr[i],
        label=f"{class_names[i]} (AUC={roc_auc[i]:.2f})"
    )

plt.plot([0,1], [0,1], "k--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves – Final Softmax")
plt.legend()
plt.savefig(os.path.join(RESULTS_DIR, "roc_curves.png"), dpi=300)
plt.close()

# ==========================================================
# STEP 11: TRAINING CURVES
# ==========================================================

plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(history.history["accuracy"], label="Train Acc")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history["loss"], label="Train Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()

plt.tight_layout()
plt.savefig(os.path.join(RESULTS_DIR, "training_curves.png"), dpi=300)
plt.close()

print("\n🎉 FINAL SOFTMAX CLASSIFIER COMPLETED SUCCESSFULLY!")
print(f"📁 All results saved at:\n{RESULTS_DIR}")
```
